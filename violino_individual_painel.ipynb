```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns


#mpl.rcParams['font.size'] = 12
#mpl.rcParams['legend.fontsize'] = 'large'
#mpl.rcParams['figure.titlesize'] = 'medium'




tabela_dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/GraÃÅficos para artigos/guara_17 (1).csv')
tabela_dados = tabela_dados.rename(columns={'hour': 'date'})
tabela_dados = pd.DataFrame(tabela_dados)

#tabela_dados = tabela_dados.drop(['date', 'NO 17', 'NO 18', 'NO 19', 'NO 20', 'NO2 17', 'NO2 18', 'NO2 19', 'NO2 20'], axis=1)
#tabela_dados = tabela_dados.drop(['date', 'NO2 17', 'NO2 18', 'NO2 19', 'NO2 20', axis=1)
#tabela_dados = tabela_dados.drop(['date', 'O3 17', 'O3 18', 'O3 19', 'O3 20', axis=1)
#tabela_dados = tabela_dados.drop('NO 17', axis=1)

tabela_dados = tabela_dados.fillna(tabela_dados.mean())


display(tabela_dados)
```

```
display(tabela_dados.describe().T)
```

```
fig, axes = plt.subplots(2, 3, figsize=(18, 10))


a = sns.violinplot(ax = axes[0,0], data = tabela_dados[['NO 17', 'NO 18', 'NO 19', 'NO 20']])
a.set_title('Guratinguet√° - NO')
a.set_ylabel('Gas concentration')
b = sns.violinplot(ax = axes[0,1], data = tabela_dados[['NO2 17', 'NO2 18', 'NO2 19', 'NO2 20']])
b.set_title('Guratinguet√° - NO2')
c = sns.violinplot(ax = axes[0,2], data = tabela_dados[['O3 17', 'O3 18', 'O3 19', 'O3 20']])
c.set_title('Guratinguet√° - O3')
d = sns.violinplot(ax = axes[1,0], data = tabela_dados[['.NO 17', '.NO 18', '.NO 19', '.NO 20']])
d.set_title('S√£o Jos√© dos Campos- NO')
d.set_ylabel('Gas concentration')
e = sns.violinplot(ax = axes[1,1], data = tabela_dados[['.NO2 17', '.NO2 18', '.NO2 19', '.NO2 20']])
e.set_title('S√£o Jos√© dos Campos- NO2')
f = sns.violinplot(ax = axes[1,2], data = tabela_dados[['.O3 17', 'O3 18.', '.O3 19', '.O3 20']])
f.set_title('S√£o Jos√© dos Campos- O3')

plt.savefig('violino.png')
```
